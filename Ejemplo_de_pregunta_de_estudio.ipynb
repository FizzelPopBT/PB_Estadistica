# Pregunta de estudio:

# ¿Existe alguna relación entre el género de los estudiantes y el área del conocimiento que eligen?

```python
import pandas as pd
df_1=pd.read_excel('Muestra_Genero_Area.xlsx')
```

```python
# Primero analizaremos la variable género
cuenta_genero=df_1.groupby('Genero').size()
tabla_genero=pd.DataFrame({
    'Frecuencia Absoluta':cuenta_genero,
    'Frecuencia Relativa %':cuenta_genero/20*100
})

tabla_genero
```

```python
#Grafico circular
import matplotlib.pyplot as plt

categorias_genero = tabla_genero.index
fr_relativa = tabla_genero['Frecuencia Relativa %']

fig,ax = plt.subplots(figsize = (6,6))
circular = ax.pie(fr_relativa, labels = categorias_genero,autopct= '%.0f%%')

#Personalización
ax.set_title('Distribución de los estudiantes según su género')

plt.show()
```

```python
#Gráfico barras
fig, ax = plt.subplots(figsize = (6,6))
colores = ['orange' if cat == 'Masculino' else 'blue' for cat in categorias_genero]
barras = ax.bar(categorias_genero, cuenta_genero, color = colores, edgecolor ='black')

ax.set_title('Distribución de los estudiantes según género')
ax.set_xlabel('Género')
ax.set_ylabel('Cantidad de estudiantes')
ax.bar_label(barras)
plt.xticks(rotation=0)
plt.show()
```

```python
# Segundo analizaremos la variable Area del conocimiento
cuenta_area=df_1.groupby('Area del conocimiento').size()
tabla_area=pd.DataFrame({
    'Frecuencia Absoluta':cuenta_area,
    'Frecuencia Relativa %':cuenta_area/20*100
})

tabla_area
```

```python
#Gráfico Circular

import matplotlib.pyplot as plt

categorias_area = tabla_area.index
fr_relativa = tabla_area['Frecuencia Relativa %']

fig,ax = plt.subplots(figsize = (6,6))
circular = ax.pie(fr_relativa, labels = categorias_area,autopct= '%.0f%%')
ax.set_title('Distribución de los estudiantes según el área del conocimiento')

plt.show()
```

```python
#Gráfico barras
fig, ax = plt.subplots(figsize = (6,6))
barras = ax.bar(categorias_area, cuenta_area, edgecolor ='black')

ax.set_title('Distribución de los estudiantes según género')
ax.set_xlabel('Género')
ax.set_ylabel('Cantidad de estudiantes')
ax.bar_label(barras)
plt.xticks(rotation=0)
plt.show()
```

```python
#Ahora analizaremos la relación entre las variables

df_relacion = df_1.groupby(["Genero","Area del conocimiento"],observed=True ).size().unstack(fill_value=0) #.unstack() = convierte un índice en columnas.
df_relacion
```

```python
# Grafico de barras de df_relacion
fig, ax = plt.subplots(figsize=(10, 6))

df_relacion.plot(kind='bar', ax=ax)

for container in ax.containers:
    ax.bar_label(container, fmt='%d')


ax.set_title('Distribución del Área del Conocimiento por Género')
ax.set_xlabel('Género')
ax.set_ylabel('Cantidad de estudiantes')
ax.tick_params(axis='x', rotation=0)
ax.legend(title='Área del conocimiento')

plt.tight_layout()
plt.show()
```

Al analizar la relación entre el género de los estudiantes y el área del conocimiento que eligen, se observa un patrón de distribución que puede reflejar ciertas tendencias socioculturales y preferencias individuales. En la muestra analizada, cada género presenta una inclinación hacia distintas áreas: por ejemplo, se puede notar que las estudiantes de género **femenino** tienden a concentrarse en áreas como **Salud y Artes**, mientras que los estudiantes de género **masculino** muestran una mayor presencia en **Tecnología e Ingeniería**.
